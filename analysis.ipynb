```python
import pandas as pd
```

```python
data=pd.read_csv('diabetes_prediction_dataset.csv')
data.head()
```

```python
x = data.drop(['diabetes'], axis = 1)
y = data['diabetes']
```

```python
y.value_counts()
```

```python
y.value_counts().plot.pie(autopct='%.2f')
```

```python
from imblearn.under_sampling import RandomUnderSampler

```

```python
rus = RandomUnderSampler(sampling_strategy = 1)
x_res, y_res = rus.fit_resample(x,y)

y_res.value_counts().plot.pie(autopct='%.2f')
```

```python
y_res.value_counts()
```

```python
data_res = pd.DataFrame(x_res, columns=x.columns)
```

```python
data_res.describe()
```

```python
y_res.describe()
```

```python
data_res.hist(bins=60, figsize=(20,10))
```

```python
cnt = len(data_res[data_res ['bmi'] > 60])
cnt
```

```python
high_bmi = data_res[data_res ['bmi'] > 60]
high_bmi
```

```python
high_bmi_df=pd.DataFrame(high_bmi, columns=['bmi'])

high_bmi_df['bmi'].plot(kind='scatter', x=high_bmi_df.index , y='bmi' )
```
